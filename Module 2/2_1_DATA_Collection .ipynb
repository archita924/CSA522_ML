HTTP response status codes 

indicate whether a specific HTTP request has been successfully completed. Responses are grouped in five classes:

Informational responses (100–199),
Successful responses (200–299),
Redirects (300–399),
Client errors (400–499),
and Server errors (500–599).

For more details visit :  https://developer.mozilla.org/en-US/docs/Web/HTTP/Status

# POST Request 

```python
# The post() method is used when you want to send some data to the server. But it can receive data too in response 
# SYNTEX 
# requests.post(url, data={key: value}, json={key: value}, args)
import requests


paylaod1={'name1': 'WWW', 'name2':'111'}
rr1=requests.post('https://httpbin.org/post',params=paylaod1)
```

```python
rr1.status_code
```

```python
rr1.url
```

```python
rr1.json() 
```

**Send image using POST request**

![title](./nitd.jpeg) 



```python
url = 'https://httpbin.org/post'
file_list = [
 ('image', ('nitd.jpeg',open('./nitd.jpeg', 'rb'), 'image/jpeg'))
] 

#use the 'auth' parameter to send requests with HTTP Basic Auth:
r_post = requests.post(url, files=file_list)
```

```python
r_post.status_code
```

```python
r_post.text
```

```python
type(r_post)
```
