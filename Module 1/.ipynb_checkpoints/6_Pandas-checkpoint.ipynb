````python
def load_with_lists():
    f = open('time_series_covid19_confirmed_global.csv','r')
    content = f.readlines()
    f.close()

    ml = []
    for x in content:
        tl = x.split(',')
        ml.append(tl)

    countries = set()
    for i in range(1,len(ml)):
        countries.add(ml[i][1])

    # for i,x in enumerate(countries):
    #     print(i,x)
    
    return countries
````

````python
f = open('time_series_covid19_confirmed_global.csv','r')
f.readlines()
````

````python
def load_with_pandas():
    import pandas as pd
    df = pd.read_csv('time_series_covid19_confirmed_global.csv')
    countries = pd.DataFrame(df['Country/Region'].unique(), columns=['Countries'])
    # print(countries)
    return countries
````

````python
df = pd.read_csv('time_series_covid19_confirmed_global.csv')
df
````

````python
import timeit

print("Lists",timeit.timeit(stmt=load_with_lists, number=100))
print("Pandas",timeit.timeit(stmt=load_with_pandas, number=100))
````

````python
#select statements
# say you want to see 
# how many people have salary greater than 3lacs, then
print("True False series (also called boolean series)\n",df['Salary'] >= 300000)
print("\nExactly those rows that satisfies the above condition are\n",df[ df['Salary'] >= 300000 ])
print("\nIf you just want the number, as to how many rows satisfies that condition\n",(df['Salary'] >= 300000).sum())
````

````python
# if you want two conditions like salary > 3lacs and gender = m then
true_false_series = (df['Salary'] >= 300000) & (df['Gender'] == 'm') #notice the round brackets around each condition, they are important
print("\nExactly those rows that satisfies the above condition are\n", df[true_false_series])
print("\nIf you just want the number, as to how many rows satisfies that condition\n",true_false_series.sum())

# another way of doing that will be
condition1_true_false_series = df['Salary'] >= 300000
condition2_true_false_series = df['Gender'] == 'm'
true_false_series = condition1_true_false_series & condition2_true_false_series
print("Exactly those rows that satisfies the above condition are\n", df[true_false_series])
print("\nIf you just want the number, as to how many rows satisfies that condition\n",true_false_series.sum())
````

````python
#datetime
import re
import datetime
def add_datepart(df, fldname, drop=True):
    df[fldname] = pd.to_datetime(df[fldname], infer_datetime_format=True)

    df['offence_day_name'] = df['Offence Date'].dt.day_name()
    df['offence_Month'] = df['Offence Date'].dt.month
    df['offence_hour'] = df['Offence Date'].dt.hour

    if drop:
        df.drop(fldname, axis=1, inplace=True)
````
